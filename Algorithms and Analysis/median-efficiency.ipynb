```python
import heapq

class MedianFinder:
    def __init__(self):
        # max_heap stores the smaller half (using negative numbers for Max-Heap behavior)
        self.max_heap = [] 
        # min_heap stores the larger half
        self.min_heap = []

    def insert(self, num):
        # 1. Decide which heap to add to
        if not self.max_heap or num <= -self.max_heap[0]:
            heapq.heappush(self.max_heap, -num)
        else:
            heapq.heappush(self.min_heap, num)

        # 2. Rebalance the heaps so size difference is at most 1
        if len(self.max_heap) > len(self.min_heap) + 1:
            # Move root of Max-Heap to Min-Heap
            val = -heapq.heappop(self.max_heap)
            heapq.heappush(self.min_heap, val)
        elif len(self.min_heap) > len(self.max_heap):
            # Move root of Min-Heap to Max-Heap
            val = heapq.heappop(self.min_heap)
            heapq.heappush(self.max_heap, -val)

    def find_median(self):
        if len(self.max_heap) == len(self.min_heap):
            return (-self.max_heap[0] + self.min_heap[0]) / 2.0
        else:
            return float(-self.max_heap[0])

    def print_heaps(self):
        # We flip the max_heap back to positive for printing
        left_side = sorted([-x for x in self.max_heap])
        right_side = sorted(self.min_heap)
        print(f"Left (Max-Heap): {left_side} | Right (Min-Heap): {right_side}")
        print(f"Current Median: {self.find_median()}\n")

# --- Example Usage ---
mf = MedianFinder()
for val in [5, 10, 2, 13]:
    print(f"Inserting {val}...")
    mf.insert(val)
    mf.print_heaps()
```
